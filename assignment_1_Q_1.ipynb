```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

from google.colab import drive

drive.mount('/content/drive')
im_1 = cv.imread('/content/drive/MyDrive/Image_Processing/im_1.jpg')
assert im_1 is not None
im_1 = cv.cvtColor(im_1, cv.COLOR_BGR2RGB)
t = np.zeros(256, dtype=np.uint8)
t[0:230] = np.array([x*255/255 for x in range(230)])
t[231:256] = np.array([x*255/255 for x in range(231, 256)])

im_2 = t[im_1]
fig, ax = plt.subplots(1, 2, figsize=(10, 10))

ax[0].imshow(im_1, vmin=0, vmax=255, cmap='gray')
ax[0].set_title('Original')
ax[1].imshow(im_2, vmin=0, vmax=255, cmap='gray')
ax[1].set_title('Intensity Transformed')

plt.show()
```
